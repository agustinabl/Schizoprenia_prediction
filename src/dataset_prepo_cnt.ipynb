## Recorrer una carpeta y leer todos los archivos con datos de sujetos Control

```python
# importamos las librerias necesarias
import os
import pandas as pd
```

```python
##############################################################################################################
# recorremos los archivos de la carpeta y los guardamos en una lista. Lo transpones y lo guardas en un dataframe.
# Luego concatenas todos los dataframes en uno solo y lo guardas en un csv.
# El csv resultante es el dataset que se usara para preprocesado posterior.
# carpeta ../data/datasets/Frontal_Aslant_Tract_L
##############################################################################################################
folder_paths = ['../data/datasets/Control/Frontal_Aslant_Tract_L']

def read_txt_to_df(file_path):
    with open(file_path, 'r') as f:
        data = f.read()
    df = pd.DataFrame([data.split('\n')])
    return df

def transpose_df(df):
    return df.transpose()

df_list = []
for folder_path in folder_paths:
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.txt'):
            file_path = os.path.join(folder_path, file_name)
            df = read_txt_to_df(file_path)
            
            df = transpose_df(df)
            df['N'] = file_name+"\t"+df.index.astype(str)
            df_list.append(df)

result = pd.concat(df_list)
result1 = result.apply(lambda x: pd.Series(x.dropna().values))
result1.to_csv('../data/union/Control/dataset_fatl.csv', index=False)
```

```python
# verificacion de los dataset
datasetcfpl
```

```python
# limpia valores nulos en la misma tabla
datasetfatl.dropna(inplace=True)
datasetfatr.dropna(inplace=True)
datasetafsl.dropna(inplace=True)
datasetafsr.dropna(inplace=True)
datasetslfl.dropna(inplace=True)
datasetslfr.dropna(inplace=True)
datasetufsl.dropna(inplace=True)
datasetufsr.dropna(inplace=True)
datasetcfpl.dropna(inplace=True)
datasetcfpr.dropna(inplace=True)
```

```python
# dividimos la columna Label en dos columnas, una con "Et" y otra con "id".
datasetfatl[['Et','id']]= datasetfatl.Label.str.split(pat='\t',expand=True)
datasetfatr[['Et','id']]= datasetfatr.Label.str.split(pat='\t',expand=True)
datasetafsl[['Et','id']]= datasetfatl.Label.str.split(pat='\t',expand=True)
datasetafsr[['Et','id']]= datasetfatr.Label.str.split(pat='\t',expand=True)
datasetslfl[['Et','id']]= datasetfatl.Label.str.split(pat='\t',expand=True)
datasetslfr[['Et','id']]= datasetfatr.Label.str.split(pat='\t',expand=True)
datasetufsl[['Et','id']]= datasetfatl.Label.str.split(pat='\t',expand=True)
datasetufsr[['Et','id']]= datasetfatr.Label.str.split(pat='\t',expand=True)
datasetcfpl[['Et','id']]= datasetfatl.Label.str.split(pat='\t',expand=True)
datasetcfpr[['Et','id']]= datasetfatr.Label.str.split(pat='\t',expand=True)
```

```python
# verificacion de los dataset
datasetfatl.head()
```

```python
# quitamos la columna Label y data de los dataset
datasetfatl.drop(columns = ['data', 'Label'], inplace=True) 
datasetfatr.drop(columns = ['data', 'Label'], inplace=True)
datasetafsl.drop(columns = ['data', 'Label'], inplace=True) 
datasetafsr.drop(columns = ['data', 'Label'], inplace=True) 
datasetslfl.drop(columns = ['data', 'Label'], inplace=True) 
datasetslfr.drop(columns = ['data', 'Label'], inplace=True) 
datasetufsl.drop(columns = ['data', 'Label'], inplace=True) 
datasetufsr.drop(columns = ['data', 'Label'], inplace=True) 
datasetcfpl.drop(columns = ['data', 'Label'], inplace=True) 
datasetcfpr.drop(columns = ['data', 'Label'], inplace=True) 
```

```python
# verificacion de los dataset
datasetcfpr.head()
```

```python
# reindexamos las columnas "Et" de los dataset
datasetfatl.reindex(columns=['Et'])
datasetfatr.reindex(columns=['Et'])
datasetafsl.reindex(columns=['Et'])
datasetafsr.reindex(columns=['Et'])
datasetslfl.reindex(columns=['Et'])
datasetslfr.reindex(columns=['Et'])
datasetufsl.reindex(columns=['Et'])
datasetufsr.reindex(columns=['Et'])
datasetcfpl.reindex(columns=['Et'])
datasetcfpr.reindex(columns=['Et'])
```

```python
# agregamos la columna "Ncodigo" a los dataset que es "codigo" con el prefijo "fatl_"
# la sigla "fatl" es por Frontal_Aslant_Tract_Left
datasetfatl['Ncodigo'] = 'fatl_'+datasetfatl['Codigo']
# agregamos la columna "Ncodigo" a los dataset que es "codigo" con el prefijo "fatl_"
# la sigla fatr es por Frontal Aslant Tract Right
datasetfatr['Ncodigo'] = 'fatr_'+datasetfatr['Codigo']

datasetafsl['Ncodigo'] = 'afsl_'+datasetafsl['Codigo']
datasetafsr['Ncodigo'] = 'afsr_'+datasetafsr['Codigo']

datasetslfl['Ncodigo'] = 'slfl_'+datasetslfl['Codigo']
datasetslfr['Ncodigo'] = 'slfr_'+datasetslfr['Codigo']

datasetufsl['Ncodigo'] = 'ufsl_'+datasetufsl['Codigo']
datasetufsr['Ncodigo'] = 'ufsr_'+datasetufsr['Codigo']

datasetcfpl['Ncodigo'] = 'cfpl_'+datasetcfpl['Codigo']
datasetcfpr['Ncodigo'] = 'cfpr_'+datasetcfpr['Codigo']

```

```python
# concatenamos los dataset
subject = pd.concat([datasetfatl, datasetfatr, 
                     datasetafsl, datasetafsr, 
                     datasetslfl, datasetslfr, 
                     datasetufsl, datasetufsr, 
                     datasetcfpl, datasetcfpr], ignore_index=True)
```

```python
# quitamos la columna "Codigo" de subject
subject.drop(columns = ['Codigo'], inplace=True) 
```

```python
# verificacion de subject
subject.head()
```

```python
#utilizamos regex para dividir la columna Et a partir de encontrar _dwi que es elimnado
# y los dos grupos de datos se agrupan en las columnas cod y scrup

subject[['cod','scrup']]= subject['Et'].str.split(r'[\_]dwi', expand=True)
```

```python
# verificacion de subject
subject.head()
```

```python
# quitamos la columna Et y scrup de subject
subject.drop(columns = ['Et','scrup'], inplace=True)
```

```python
# reindexamos la columna cod de subject
subject.reindex(columns=['cod'])
```

```python
# verificacion de subject
subject.head()
```

```python
# pivotamos la tabla subject para que los valores de la columna Ncodigo se conviertan en columnas
df = subject.pivot(index=['cod'], columns='Ncodigo', values='values').reset_index()
```
